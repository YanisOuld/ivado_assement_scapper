# Museum Attendance vs City Population

This notebook explores the relationship between:
- annual museum attendance
- city population

Data sources:
- Wikipedia: list of most visited museums
- Kaggle: world cities population dataset

All data processing is handled in the `src/` pipeline.
This notebook focuses on exploration, visualization, and interpretation.

```python
import sys
from pathlib import Path

sys.path.append(str(Path("..").resolve()))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
museum_df = pd.read_csv("../data/raw/museums_raw.csv")
museum_df["annual_visitors"] = museum_df["annual_visitors"].astype(float)
museum_df
```

```python
plt.bar(museum_df["museum_name"], museum_df["annual_visitors"])
plt.xlabel("museum_name")
plt.ylabel("Number of visitor")
plt.title("Attendance of museum for each contry")
plt.show()
```

```python
dataset = pd.read_csv("../data/processed/dataset.csv")
plt.scatter(dataset["annual_visitors"], dataset["city_population"])
plt.xlabel("Visitors")
plt.ylabel("Population")
plt.title("Correlation")
plt.show()
```

```python
plt.scatter(dataset["log_annual_visitors"], dataset["log_city_population"])
plt.xlabel("Logged visitors")
plt.ylabel("Logged population")
plt.title("Correlation of the log")
plt.show()
```
